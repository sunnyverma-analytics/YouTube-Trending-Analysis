**Visualisation**

**Matplotlib**


**Bar Chart**

```python
import matplotlib.pyplot as plt
```

```python
top10 = df.groupby('channel_title')['views'].sum() \
          .sort_values(ascending=False) \
          .head(10)
top10
```

```python
import matplotlib.pyplot as plt

top10 = df.groupby('channel_title')['views'].sum() \
          .sort_values(ascending=False) \
          .head(10)

plt.figure(figsize=(12,6))

top10.plot(
    kind='bar',
    color=['red','blue','green','orange','purple',
           'brown','pink','gray','cyan','gold']
)

plt.title('Top 10 Channels by Total Views')
plt.xlabel('Channel Title')
plt.ylabel('Total Views')

plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

**Category-wise Average Views**

```python
category_avg = df.groupby('category_id')['views'].mean()
category_avg.plot(kind='bar')
```

**Likes vs Views**

```python
plt.scatter(df['views'], df['likes'])
plt.xlabel('Views')
plt.ylabel('Likes')
plt.show()
```

**Business Insights**

**Identify which content category contributes the highest share of views.**

```python
top_category = df.groupby('category_id')['views'].sum().sort_values(ascending=False)
top_category.head()
```

**Identify channels that dominate audience attention.**

```python
top_channel = df.groupby('channel_title')['views'].sum().sort_values(ascending=False)
top_channel.head(10)
```
